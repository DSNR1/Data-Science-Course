# **Seaborn**

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#tips=sns.load_dataset("tips.csv")
tips=mtcars
```

```python
tips.head()
```

# **Univariate ==> "One variable" data visualization**

```python
#Strip plot
sns.stripplot(y='qsec',data=tips,jitter=True)
plt.ylabel('tip($)')
plt.show()
```

```python
#Grouping with stripplot()
sns.stripplot(x='gear',y='qsec',data=tips)
plt.show()
```

```python
sns.stripplot(x='gear',y='qsec',data=tips,size=4,jitter=False)
plt.ylabel('Tip($)')
plt.show()
```

```python
#Swarm plot
sns.swarmplot(x="gear", y="qsec", data = tips)
plt.ylabel("tips")
plt.show()
```

```python
sns.swarmplot(x="gear", y="qsec", data = tips, hue='cyl')
plt.ylabel("tips")
plt.show()
```

```python
#Changing the orientation 
sns.swarmplot(x="gear", y="qsec", data = tips, hue='cyl', orient='h')
plt.ylabel("tips")
plt.show()
```

```python
sns.boxplot(y='qsec',data=tips)
plt.ylabel('Tips')
```

# **Box and Violin plot**

```python
sns.boxplot(x='gear',y='qsec',data=tips)
plt.ylabel('Tips')
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='gear',y='qsec',data=tips)
plt.ylabel('Tips')
```

```python
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('Tips')
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='gear',y='qsec',data=tips)
plt.ylabel("Tips")
plt.subplot(1,2,2)
sns.violinplot(x='cyl',y='qsec',data=tips)
plt.ylabel("Tips")
plt.show()
```

```python
#Combinining plots
sns.violinplot(x='gear', y='qsec',data=tips, inner=None ,  color='lightgray')
sns.stripplot(x='gear', y='qsec',data=tips, size=4, jitter=True)
plt.ylabel("Tip")
plt.show()
```

# **Bivariate ==> "Two variable" data visualization**

```python
#Joint plot
sns.jointplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
#Density plot
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde")  #Kernal density estimate
plt.show()
```

```python
#Pair plot
sns.pairplot(tips)
plt.show()
```

```python
#Pair plot
sns.pairplot(tips, hue="sex")
plt.show()
```

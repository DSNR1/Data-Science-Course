# Fitting a Linear Regression Model

```python
import statsmodels.formula.api as smf
model= smf.ols("AT~Waist", data = data).fit()
```

```python
sns.regplot(x= "Waist", y="AT", data=data);
```

```python
# Coefficients
model.params
```

```python
# t and p-values
print(model.tvalues,'\n', model.pvalues)
```

```python
# Rsquared Values
(model. rsquared)
```
